# Get review data and teams keywords

```python
import pandas as pd
import os
from tqdm import tqdm
import json
```

```python
REVIEW_PATH = '../Resources/1_clean_review_data'

KEYWORDS_PATH = '../Resources/2_keywords_for_teams_data'

DATA_PATH = '../Resources'

REVIEW_COL = 'Review'
TITLE_COL = 'Title'
```

```python
data = pd.read_csv(os.path.join(REVIEW_PATH,'cleanAppStore.csv'))
```

```python
with open(os.path.join(KEYWORDS_PATH,'team_keywords.json'),'r') as f:
    team_names_keywords = json.load(f)

with open(os.path.join(KEYWORDS_PATH,'grouped_team_keywords.json'),'r') as f:
    groups_keywords = json.load(f)
    
with open(os.path.join(KEYWORDS_PATH,'grouped_team_mapping.json'),'r') as f:
    mapping = json.load(f)
```

```python
team_names_keywords_vects = [' '.join(team_names_keywords[team]) for team in team_names_keywords]
```

# Text preprocessing all words

```python
"""removes punctuation, stopwords, and returns a list of the remaining words, or tokens"""
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import string
def text_process(text):
    '''
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Return the cleaned text as a list of words
    4. Remove words
    '''
    stemmer = WordNetLemmatizer()
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join([i for i in nopunc if not i.isdigit()])
    nopunc =  [word.lower() for word in nopunc.split() if word.lower() not in stopwords.words('russian')]
    return ' '.join([stemmer.lemmatize(word) for word in nopunc])
```

```python
preprocessed_reviews = []
for i in tqdm(data.index):
    text = data.iloc[i][REVIEW_COL]+' '+data.iloc[i][TITLE_COL]
    preprocessed_reviews.append(text_process(text))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfconvert = TfidfVectorizer().fit(preprocessed_reviews)

X_transformed=tfidfconvert.transform(preprocessed_reviews)
```

```python
tfidfconvert.vocabulary_
```

```python
X_transformed.shape
```

# K-Mean clustering for all words

```python
from sklearn.cluster import KMeans

Sum_of_squared_distances = []
K = range(1,50)
for k in tqdm(K):
    km = KMeans(n_clusters=k)
    km = km.fit(X_transformed)
    Sum_of_squared_distances.append(km.inertia_)


import matplotlib.pyplot as plt

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
km = KMeans(n_clusters=20)
```

```python
km = km.fit(X_transformed)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2, random_state=2)
reduced_features = pca.fit_transform(X_transformed.toarray())
```

```python
reduced_features
```

```python
reduced_cluster_centers = pca.transform(km.cluster_centers_)
```

```python
plt.scatter(reduced_features[:,0], reduced_features[:,1], c=km.predict(X_transformed))
plt.scatter(reduced_cluster_centers[:, 0], reduced_cluster_centers[:,1], marker='x', s=15, c='b')
```

```python
teams_transformed = tfidfconvert.transform(team_names_keywords_vects)
```

```python
df_clusters = pd.DataFrame(columns=['team','cluster'])
```

```python
df_clusters['team'] = team_names_keywords.keys()
df_clusters['cluster'] = km.predict(teams_transformed)
```

```python
from collections import Counter
counter = Counter(km.labels_)
for cluster_num in range(20):
    team_in_cluster = df_clusters[df_clusters['cluster']==cluster_num]['team']
    if len(team_in_cluster)!=0:
        print ('Teams in cluster:',cluster_num,'; '.join(team_in_cluster))
        print('Reviews in cluster:', counter[cluster_num])
    
```

# Text preprocessing key words

```python
import string
def text_process(text):
    '''
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Return the cleaned text as a list of words
    4. Remove words
    '''
    
    from rutermextract import TermExtractor
    term_extractor = TermExtractor()
    return ' '.join([term.normalized for term in term_extractor(text)])
```

```python
preprocessed_reviews_key = []
for i in tqdm(data.index):
    text = data.iloc[i][REVIEW_COL]+' '+data.iloc[i][TITLE_COL]
    preprocessed_reviews_key.append(text_process(text))
```

```python
for i, elem in enumerate(preprocessed_reviews_key):
    new_elem = []
    for word in elem:
        new_elem.append('_'.join(word.split(' ')))
    preprocessed_reviews_key[i] = new_elem
        
```

```python
preprocessed_reviews_key
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfconvert = TfidfVectorizer().fit(preprocessed_reviews_key)

X_transformed=tfidfconvert.transform(preprocessed_reviews_key)
```
