# Compute quality metrics

```python
LABEL_COL = 'label'
```

```python
data_marked_df = pd.read_csv(os.path.join(DATA_PATH,'marked.csv'),index_col = 0)
```

```python
y_true = [y.split(';') for y in data_marked_df[LABEL_COL]]
```

```python
bert_eucl_preds = list(data_marked_df['bert_euclid_team'].values)
bert_cos_preds = list(data_marked_df['bert_cosine_team'].values)
```

```python
flair_eucl_preds = list(data_marked_df['flair_euclid_team'].values)
flair_cos_preds = list(data_marked_df['flair_cosine_team'].values)
```

```python
tfidf_cos_preds = list(data_marked_df['tfidf_cos_preds'].values)
```

```python
keywords_preds = [y[2:-2].split('\', \'')[0] for y in data_marked_df['keywords_preds'].values]
# keywords_preds = [preds if len(preds)!=1 and preds[0]!='' else ['iOS Platform'] for preds in keywords_preds]
```

```python
metrics = []
```

```python
def accuracy(y_trues,  y_preds):
    import numpy as np
    accs = []
    for j in range(3):
        y_true = y_trues[j*100:(j+1)*100]
        y_pred = y_preds[j*100:(j+1)*100]


        acc = 0
        for i,pred in enumerate(y_pred):
            if pred in y_true[i]:
                acc+=1
            else:
                pass
        accs.append(acc/len(y_pred))
    return np.mean(accs), np.std(accs)
```

```python
mean, std = accuracy(y_true,bert_eucl_preds)
print('Bert eucleadean metric accuracy = ',round(mean,2),'+/-',round(std,2))
metrics.append({'model':'BERT with euclidean distance','accuracy_mean':mean, 'accuracy_std':std})
mean, std = accuracy(y_true,bert_cos_preds)
print('Bert cosine metric accuracy = ',round(mean,2),'+/-',round(std,2))
metrics.append({'model':'BERT with cosine distance','accuracy_mean':mean, 'accuracy_std':std})
print()
mean, std = accuracy(y_true,flair_eucl_preds)
print('Flair eucleadean metric accuracy = ',round(mean,2),'+/-',round(std,2))
metrics.append({'model':'Flair with euclidean distance','accuracy_mean':mean, 'accuracy_std':std})
mean, std = accuracy(y_true,flair_cos_preds)
print('Flair cosine metric accuracy = ',round(mean,2),'+/-',round(std,2))
metrics.append({'model':'Flair with cosine distance','accuracy_mean':mean, 'accuracy_std':std})
print()
mean, std = accuracy(y_true,tfidf_cos_preds)
print('TFIDF cosine metric accuracy = ',round(mean,2),'+/-',round(std,2))
metrics.append({'model':'TFIDF with cosine distance','accuracy_mean':mean, 'accuracy_std':std})
print()
mean, std = accuracy(y_true,keywords_preds)
print('Keywords matching accuracy = ',round(mean,2),'+/-',round(std,2))
metrics.append({'model':'Keywords matching','accuracy_mean':mean, 'accuracy_std':std})

```

```python
def accuracy_wo_platform(y_trues,  y_preds):
    import numpy as np
    accs = []
    for j in range(3):
        y_true = y_trues[j*100:(j+1)*100]
        y_pred = y_preds[j*100:(j+1)*100]


        acc = 0
        for i,pred in enumerate(y_pred):
            if pred in y_true[i] and pred!='iOS Platform':
                acc+=1
            else:
                pass
        accs.append(acc/len(y_pred))
    return np.mean(accs), np.std(accs)
```

```python
mean, std = accuracy_wo_platform(y_true,bert_eucl_preds)
print('Bert eucleadean metric accuracy = ',round(mean,2),'+/-',round(std,2))
mean, std = accuracy_wo_platform(y_true,bert_cos_preds)
print('Bert cosine metric accuracy = ',round(mean,2),'+/-',round(std,2))
print()
mean, std = accuracy_wo_platform(y_true,flair_eucl_preds)
print('Flair eucleadean metric accuracy = ',round(mean,2),'+/-',round(std,2))
mean, std = accuracy_wo_platform(y_true,flair_cos_preds)
print('Flair cosine metric accuracy = ',round(mean,2),'+/-',round(std,2))
print()
mean, std = accuracy_wo_platform(y_true,tfidf_cos_preds)
print('TFIDF cosine metric accuracy = ',round(mean,2),'+/-',round(std,2))
print()
mean, std = accuracy_wo_platform(y_true,keywords_preds)
print('Keywords matching accuracy = ',round(mean,2),'+/-',round(std,2))
```

```python
def recall(y_trues,  y_preds):
    classes = list(team_names_keywords.keys())+['']
    import numpy as np
    from collections import Counter
    recalls = []
    for j in range(3):
        
        tp = np.zeros(len(classes))
        fp = np.zeros(len(classes))
        
        y_true = y_trues[j*100:(j+1)*100]
        y_pred = y_preds[j*100:(j+1)*100]
        
        for i,pred in enumerate(y_pred):
            if pred in y_true[i]:
                tp[classes.index(pred)]+=1
            else:
                for true in y_true[i]:
                    try:
                        fp[classes.index(true.rstrip('\'').rstrip('.').lstrip('"'))]+=1
                    except ValueError:
                        pass
        recalls.append(np.sum(tp)/np.sum(tp+fp))
    return np.mean(recalls), np.std(recalls)

```

```python
def prescision(y_trues,  y_preds):
    import numpy as np
    from collections import Counter
    recalls = []
    classes = list(team_names_keywords.keys())+[np.nan]+['']


    for j in range(3):
        
        tp = np.zeros(len(classes))
        fp = np.zeros(len(classes))
        
        y_true = y_trues[j*100:(j+1)*100]
        y_pred = y_preds[j*100:(j+1)*100]
        
        for i,pred in enumerate(y_pred):
            if pred in y_true[i]:
                tp[classes.index(pred)]+=1
            else:
                fp[classes.index(pred)]+=1
        recalls.append(np.sum(tp)/np.sum(tp+fp))
    return np.mean(recalls), np.std(recalls)


```

```python
mean, std = recall(y_true,bert_eucl_preds)
print('Bert eucleadean metric recall = ',round(mean,2),'+/-',round(std,2))
metrics[0]['recall_mean']  = mean
metrics[0]['recall_std']  = std
mean, std = recall(y_true,bert_cos_preds)
print('Bert cosine metric recall = ',round(mean,2),'+/-',round(std,2))
metrics[1]['recall_mean']  = mean
metrics[1]['recall_std']  = std
print()
mean, std = recall(y_true,flair_eucl_preds)
print('Flair eucleadean metric recall = ',round(mean,2),'+/-',round(std,2))
metrics[2]['recall_mean']  = mean
metrics[2]['recall_std']  = std
mean, std = recall(y_true,flair_cos_preds)
print('Flair cosine metric recall = ',round(mean,2),'+/-',round(std,2))
metrics[3]['recall_mean']  = mean
metrics[3]['recall_std']  = std
print()
mean, std = recall(y_true,tfidf_cos_preds)
print('TFIDF cosine metric recall = ',round(mean,2),'+/-',round(std,2))
metrics[4]['recall_mean']  = mean
metrics[4]['recall_std']  = std
print()
mean, std = recall(y_true,keywords_preds)
print('Keywords matching recall = ',round(mean,2),'+/-',round(std,2))
metrics[5]['recall_mean']  = mean
metrics[5]['recall_std']  = std
```

```python
mean, std = prescision(y_true,bert_eucl_preds)
print('Bert eucleadean metric prescision = ',round(mean,2),'+/-',round(std,2))
mean, std = prescision(y_true,bert_cos_preds)
print('Bert cosine metric prescision = ',round(mean,2),'+/-',round(std,2))
print()
mean, std = prescision(y_true,flair_eucl_preds)
print('Flair eucleadean metric prescision = ',round(mean,2),'+/-',round(std,2))
mean, std = prescision(y_true,flair_cos_preds)
print('Flair cosine metric prescision = ',round(mean,2),'+/-',round(std,2))
print()
mean, std = prescision(y_true,tfidf_cos_preds)
print('TFIDF cosine metric prescision = ',round(mean,2),'+/-',round(std,2))
print()
mean, std = prescision(y_true,keywords_preds)
print('Keywords matching prescision = ',round(mean,2),'+/-',round(std,2))
```

```python
def f1_score(y_trues, y_preds):
    
    import numpy as np
    from collections import Counter
    classes = list(team_names_keywords.keys())+['']+[np.nan]
    f1s = []
    for j in range(10):
        
        tp = np.zeros(len(classes))
        fp = np.zeros(len(classes))
        fn = np.zeros(len(classes))
        
        y_true = y_trues[j*10:(j+1)*10]
        y_pred = y_preds[j*10:(j+1)*10]
        
        for i,pred in enumerate(y_pred):
            if pred in y_true[i]:
                tp[classes.index(pred)]+=1
            else:
                fp[classes.index(pred)]+=1
                for true in y_true[i]:
                    try:
                        fn[classes.index(true.rstrip('\''))]+=1
                    except ValueError:
                        pass
        recall = np.sum(tp)/np.sum(tp+fn)
        precision = np.sum(tp)/np.sum(tp+fp)
        if recall==0 and precision==0:
            continue


        f1s.append(2*precision*recall/(precision+recall))
    return np.mean(f1s), np.std(f1s)
```

```python
mean, std = f1_score(y_true,bert_eucl_preds)
print('Bert eucleadean metric f1 = ',round(mean,2),'+/-',round(std,2))
metrics[0]['f1_score_mean']  = mean
metrics[0]['f1_score_std']  = std

mean, std = f1_score(y_true,bert_cos_preds)
print('Bert cosine metric f1 = ',round(mean,2),'+/-',round(std,2))

metrics[1]['f1_score_mean']  = mean
metrics[1]['f1_score_std']  = std
print()
mean, std = f1_score(y_true,flair_eucl_preds)
print('Flair eucleadean metric f1 = ',round(mean,2),'+/-',round(std,2))
metrics[2]['f1_score_mean']  = mean
metrics[2]['f1_score_std']  = std
mean, std = f1_score(y_true,flair_cos_preds)
print('Flair cosine metric f1 = ',round(mean,2),'+/-',round(std,2))
metrics[3]['f1_score_mean']  = mean
metrics[3]['f1_score_std']  = std
print()
mean, std = f1_score(y_true,tfidf_cos_preds)
print('TFIDF cosine metric f1 = ',round(mean,2),'+/-',round(std,2))
metrics[4]['f1_score_mean']  = mean
metrics[4]['f1_score_std']  = std
print()
mean, std = f1_score(y_true,keywords_preds)
print('Keywords matching f1 = ',round(mean,2),'+/-',round(std,2))
metrics[5]['f1_score_mean']  = mean
metrics[5]['f1_score_std']  = std
```
